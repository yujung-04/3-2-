```python
import pandas as pd

train_df = pd.read_csv("/content/train_categorized.csv", encoding = "utf-8-sig")
test_df = pd.read_csv("/content/test_categorized.csv", encoding = "utf-8-sig")
val_df = pd.read_csv("/content/val_categorized.csv", encoding = "utf-8-sig")

train_df
```

```python
df = pd.concat([train_df, test_df, val_df], axis=0).reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.35, shuffle=True, random_state=42)
```

```python
train["text"] = train["title"].fillna("") + " " + train["content"].fillna("")
test["text"]  = test["title"].fillna("")  + " " + test["content"].fillna("")

# 확인
train[["title", "content", "text"]].head()
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# 텍스트 전처리: 제목과 본문을 합쳐 하나의 입력으로 구성
train["text"] = train["title"].fillna("") + " [SEP] " + train["content"].fillna("")
test["text"]  = test["title"].fillna("")  + " [SEP] " + test["content"].fillna("")

# 최소 개수 기준으로 클래스 필터링 (데이터 불균형 완화 목적)
MIN_COUNT = 20
cls_counts = train["category"].value_counts()
valid_classes = cls_counts[cls_counts > MIN_COUNT].index.tolist()

train_filtered = train[train["category"].isin(valid_classes)]
test_filtered  = test[test["category"].isin(valid_classes)]

print("사용되는 클래스:", valid_classes)
print("학습 데이터 수:", len(train_filtered))
print("클래스별 개수:\n", train_filtered["category"].value_counts())
```

```python
# SBERT 임베딩 생성: 한국어 멀티태스크 RoBERTa 모델 사용
from sentence_transformers import SentenceTransformer

MODEL_NAME = "jhgan/ko-sroberta-multitask"
embedder = SentenceTransformer(MODEL_NAME)

X_train = embedder.encode(train_filtered["text"].tolist(),
                          batch_size=16, convert_to_numpy=True)
X_test  = embedder.encode(test_filtered["text"].tolist(),
                          batch_size=16, convert_to_numpy=True)

y_train = train_filtered["category"].tolist()
y_test  = test_filtered["category"].tolist()

# SMOTE로 소수 클래스 샘플을 생성해 데이터 불균형 보정
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# 표준화(정규화): 평균 0, 표준편차 1로 변환해 MLP 학습 안정화
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_res)
X_test_scaled  = scaler.transform(X_test)

# MLP 분류기: 깊은 은닉층 구조 + early stopping으로 과적합 방지
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(
    hidden_layer_sizes=(768, 512, 256),
    activation='relu',
    solver='adam',
    learning_rate_init=1e-4,
    max_iter=30000,
    batch_size=64,
    early_stopping=True,
    n_iter_no_change=20,
    validation_fraction=0.1,
    random_state=42
)

clf.fit(X_train_scaled, y_train_res)

# 모델 평가: Accuracy 및 클래스별 성능 출력
from sklearn.metrics import accuracy_score, classification_report

test_pred = clf.predict(X_test_scaled)

print("Test Accuracy:", accuracy_score(y_test, test_pred))
print("Test Report:\n", classification_report(y_test, test_pred))
```
